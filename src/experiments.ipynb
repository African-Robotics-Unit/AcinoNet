## Parewise-Extrinsic Calibration with Dummy Scene Data

```python
import os
import sys
import glob
from pprint import pprint
import numpy as np
import cv2
sys.path.append('./Calib')
from calib import calib, app


DATA_ROOT_DIR = '../data'
CONFIG_DIR = os.path.abspath('../configs')
INTRINSIC_DATA_DIR = os.path.join(DATA_ROOT_DIR, 'intrinsic_calibration')


NUM_CAMERA = 6

# generate dir paths including extrinsic calibration videos
day_paths = glob.glob(os.path.join(DATA_ROOT_DIR, '??_??_20??'))
extrinsic_data_paths = []
for day_path in day_paths:
    extrinsic_data_paths.append(
        os.path.join(day_path, 'extrinsic_calibration')
    )
# pprint(extrinsic_data_paths)

extrinsic_data_path = extrinsic_data_paths[2]
# print(extrinsic_data_path)


# Get paths of the calibration videos
calib_video_path = os.path.join(extrinsic_data_path, 'videos')
calib_vids = sorted(glob.glob(os.path.join(calib_video_path, '*.mp4')))
assert len(calib_vids) == NUM_CAMERA, 'expect {} camera videos but the number of videos is {}'.format(
    NUM_CAMERA, len(calib_vids)
)
# pprint(calib_vids)



camera_fpaths = [os.path.join(INTRINSIC_DATA_DIR, "camera.json") for _ in range(NUM_CAMERA)]
points_fpaths = [os.path.join(extrinsic_data_path, f"points/points_{i+1}.json") for i in range(NUM_CAMERA)]
scene_fpath = os.path.join(extrinsic_data_path, "scene.json")
app.calibrate_fisheye_extrinsics_pairwise(
    camera_fpaths, points_fpaths, 
    out_fpath=scene_fpath,
    dummy_scene_path=os.path.join(CONFIG_DIR, 'dummy_scene.json'),
    start_camera_i=0
)
```

## SBA with Pre-Difined Points

```python
import os
import sys
import glob
from pprint import pprint
import numpy as np
np.set_printoptions(threshold=sys.maxsize)
import cv2
sys.path.append('./Calib')
from calib import calib, app


DATA_ROOT_DIR = '../data'
CONFIG_DIR = os.path.abspath('../configs')
INTRINSIC_DATA_DIR = os.path.join(DATA_ROOT_DIR, 'intrinsic_calibration')


NUM_CAMERA = 6

# generate dir paths including extrinsic calibration videos
day_paths = glob.glob(os.path.join(DATA_ROOT_DIR, '??_??_20??'))
extrinsic_data_paths = []
for day_path in day_paths:
    extrinsic_data_paths.append(
        os.path.join(day_path, 'extrinsic_calibration')
    )
# pprint(extrinsic_data_paths)

extrinsic_data_path = extrinsic_data_paths[2]
# print(extrinsic_data_path)


# Get paths of the calibration videos
calib_video_path = os.path.join(extrinsic_data_path, 'videos')
calib_vids = sorted(glob.glob(os.path.join(calib_video_path, '*.mp4')))
assert len(calib_vids) == NUM_CAMERA, 'expect {} camera videos but the number of videos is {}'.format(
    NUM_CAMERA, len(calib_vids)
)
# pprint(calib_vids)



camera_fpaths = [os.path.join(INTRINSIC_DATA_DIR, "camera.json") for _ in range(NUM_CAMERA)]
points_fpaths = [os.path.join(extrinsic_data_path, f"points/points_{i+1}.json") for i in range(NUM_CAMERA)]
scene_fpath = os.path.join(extrinsic_data_path, "scene.json")



defined_point_fpath = os.path.join(extrinsic_data_path, f"defined_points/defined_points.json")
scene_sba_fpath = os.path.join(extrinsic_data_path, "scene_sba.json")
res = app.sba_board_points_fisheye(
    scene_fpath, points_fpaths, defined_point_fpath,
    out_fpath=scene_sba_fpath
)
print(f"Before: mean: {np.mean(res['before'])}, std: {np.std(res['before'])}")
print(f"After: mean: {np.mean(res['after'])}, std: {np.std(res['after'])}")
```
